# P value
The p value is a number, calculated from a statistical test, that describes how likely you are to have found a particular set of observations if the null hypothesis were true. P values are used in hypothesis testing to help decide whether to reject the null hypothesis


 It helps scientists decide if the results they've found are just due to chance or if they're actually meaningful.

Here's a simple way to understand it:

Imagine you're flipping a coin, and you want to know if it's a fair coin (meaning it has an equal chance of landing on heads or tails). You flip the coin 10 times and get 7 heads and 3 tails. Now, you want to know if this result is just a fluke or if it suggests the coin is biased toward heads.

The p-value in this case tells you how likely it is to get 7 heads and 3 tails (or something even more extreme) just by random chance if the coin is truly fair. If the p-value is very low, like 0.05 or less, it suggests that your result is quite unusual and not likely to happen by chance. So, you might conclude that the coin is probably not fair.

On the other hand, if the p-value is high, like 0.5 or more, it suggests that getting 7 heads and 3 tails is quite common and could easily happen with a fair coin. In this case, you might conclude that there's no strong evidence that the coin is biased.

So, in essence, a p-value helps scientists decide if their findings are statistically significant, meaning they're likely to be real and not just the result of random luck. A lower p-value indicates stronger evidence against the null hypothesis (e.g., the coin is fair), while a higher p-value suggests that the data doesn't provide strong evidence to reject the null hypothesis. Typically, scientists use a threshold (often 0.05) to determine whether a p-value is considered statistically significant or not.

# Confidence interval

confidence interval is like a range or a band of values that helps us estimate where a particular number or measurement is likely to be. Imagine you're trying to guess the height of a giraffe in a zoo, but you can't measure it directly. Instead, you use a tool that gives you a range of possible heights. This range is your "confidence interval."

Here's how it works:

Let's say you measure the giraffe's height and find it to be 16 feet, but you're not 100% certain because your measuring tool may have a bit of error. So, you calculate a 95% confidence interval, which might be something like 15.5 feet to 16.5 feet. This means you're pretty confident that the giraffe's height falls within this range.

If you calculated a 90% confidence interval instead, it might be slightly wider, like 15.3 feet to 16.7 feet. This means you're a bit more uncertain, but you still think the giraffe's height is in this range.

# Confidence Interval Formula

The confidence interval is based on the mean and standard deviation. Thus, the formula to find CI is

X̄ ± Zα/2 × [ σ / √n ]

Where

X̄ = Mean

Z = Confidence coefficient

α = Confidence level

σ = Standard deviation

N = sample space

The value after the ± symbol is known as the margin of error.

# Confidence Interval Example

Question: In a tree, there are hundreds of apples. You are randomly choosing 46 apples with a mean of 86 and a standard deviation of 6.2. Determine that the apples are big enough.

Solution:

Given: Mean, X̄ = 86

Standard deviation, σ = 6.2

Number of observations, n = 46

Take the confidence level as 95%. Therefore, the value of z = 1.960 (from the table)

The formula to find the confidence interval is

X̄ ± Zα/2 × [ σ / √n ]

Now, substitute the values in the formula, we get

86 ± 1.960 × [ 6.2 / √46 ]

86 ± 1.960 × [ 6.2 / 6.78]

86 ± 1.960 × 0.914

86 ± 1.79

Here, the margin of error is 1.79

Therefore, all the hundreds of apples are likely to be between in the range of 84. 21 and 87.79.#

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some example data (you can replace this with your own dataset)
np.random.seed(42)  # Setting a seed for reproducibility
data = np.random.normal(loc=5, scale=2, size=100)  # Mean=5, Standard Deviation=2, 100 data points
#data = [78,65,20,69,36,81,85,71,44,71,47,37,89,25,73,55,80,52,46,77]
# Calculate the confidence interval formulaically
confidence_level = 0.95  # 95% confidence interval
data_mean = np.mean(data)
data_std = np.std(data)
sample_size = len(data)
margin_of_error = 1.96 * (data_std / np.sqrt(sample_size))  # For a 95% confidence interval (z-value for 95% CI)

# Calculate the confidence interval bounds
confidence_interval_lower = data_mean - margin_of_error
confidence_interval_upper = data_mean + margin_of_error

# Plot the data and confidence interval
sns.distplot(data, bins=20, color='b', label='Random Data')

plt.axvline(data_mean, color='r', linestyle='--', label='Mean')
plt.axvline(confidence_interval_lower, color='g', linestyle='-', label='Confidence Interval (Lower)')
plt.axvline(confidence_interval_upper, color='g', linestyle='-', label='Confidence Interval (Upper)')

# Fill the area between the confidence interval bounds with a color
#plt.fill_betweenx([0, 25], confidence_interval_lower, confidence_interval_upper, color='gray', alpha=0.5, label='Confidence Interval')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Random Data with Confidence Interval')
plt.show()

# Print the confidence interval
print(f'Confidence Interval ({confidence_level * 100}%): ({confidence_interval_lower}, {confidence_interval_upper})')
```

# Decision Rules

The two methods of concluding the Hypothesis test are using the Test-statistic value, p-value.

In both methods, we start assuming the Null Hypothesis to be true, and then we reject the Null hypothesis if we find enough evidence.

The decision rule for the Test-statistic method:

if test-statistic (t) > critical Value (C), we reject Null Hypothesis.

If test-statistic (t) ≤ critical value (C), we fail to reject Null Hypothesis.

The decision rule for the p-value method:

if p-value (p) > level of significance (α), we fail to reject Null Hypothesis

if p-value (p) ≤ level of significance (α), we reject Null Hypothesis

# Errors in Hypothesis Testing

# Type 1 Error

The Type 1 error is the false positive error where we have rejected the null hypothesis but it is actually true.  

# Type 2 Error

Type 2 error is a false negative conclusion where we have not rejected the null hypothesis but it is actually false.

#One-Tailed vs Two-Tailed T-Test

# One tail

A statistical hypothesis test in which alternative hypothesis has only one end, is known as one tailed test

# Two tail

A significance test in which alternative hypothesis has two ends, is called two-tailed test.

The concepts of one-tailed and two-tailed t-tests are important in the realm of hypothesis testing, particularly when using the t-distribution to evaluate the significance of a set of data relative to a theoretical model. These tests help determine whether to reject the null hypothesis, based on the data collected and the nature of the hypothesis being tested.

### One-Tailed t-Test

A one-tailed t-test is used when the research hypothesis (alternative hypothesis) specifies a direction of the effect. This means that the investigator predicts not only a difference between the groups but also the direction of this difference.

**Examples:**
- Testing if the mean of a single group is greater than a specific value.
- Testing if the mean of one group is greater than the mean of another group.

**Hypothesis Setup:**
- **Right-Tailed Test**: Here, you test for an increase or improvement. For example,
  - \( H_0: \mu = \mu_0 \)
  - \( H_a: \mu > \mu_0 \)
- **Left-Tailed Test**: Here, you test for a decrease or reduction. For example,
  - \( H_0: \mu = \mu_0 \)
  - \( H_a: \mu < \mu_0 \)

In these cases, the critical region for the test statistic is entirely in one tail of the distribution (either the right or the left).

### Two-Tailed t-Test

A two-tailed t-test is used when the research hypothesis does not specify a direction of the effect. This test is appropriate when you are looking for any difference from the theoretical value, regardless of the direction of this difference.

**Example:**
- Testing if the mean of a group is different from a specific value (could be either higher or lower).
- Testing if the means of two groups are different (irrespective of which is higher).

**Hypothesis Setup:**
- \( H_0: \mu = \mu_0 \)
- \( H_a: \mu \neq \mu_0 \)

Here, the critical region is split between both tails of the distribution, accounting for extremes in either direction.

### Calculating and Interpreting t-Tests

In practice, the t-test involves calculating a test statistic, which is then compared against values from the t-distribution. The steps are:

1. **Calculate the t-statistic** from your sample data. This involves the mean of the sample, the standard deviation of the sample, and the sample size.
2. **Determine the critical t-value(s)** from t-distribution tables (or software), based on your chosen significance level (\(\alpha\)) and degrees of freedom (which generally correspond to \(n - 1\) for a one-sample t-test, where \(n\) is the sample size).
3. **Compare your calculated t-statistic** to the critical t-value:
   - For a one-tailed test, compare to one critical value in the specified tail.
   - For a two-tailed test, compare to critical values in both tails (typically, you take the positive and negative of the same critical value).

**Decision Making:**
- If the t-statistic falls within the critical region (beyond the critical value(s)), reject the null hypothesis.
- If the t-statistic does not fall within the critical region, do not reject the null hypothesis.

### Considerations

- **Choice Between One-Tailed and Two-Tailed**: The choice depends on your research question. A one-tailed test has more power to detect an effect in one direction but risks missing an effect in the other direction.
- **Significance Level**: Commonly set at 0.05, this is the probability of rejecting the null hypothesis when it is actually true (Type I error).

In summary, the decision between using a one-tailed or two-tailed t-test hinges on the nature of your hypothesis and what you aim to discover or confirm with your data.

```python
#Q The average height of Indian adult males is 165cm.

#One sample t test

#Null hypothesis: The average height is 165cm.
#Alternate Hypothesis: The average height is not 165cm.

import numpy as np
from scipy.stats import ttest_1samp
from random import sample

# generating a random sample to get the heights

sample1 = sample(range(145,180), 30)

#calculation sample mean

sample_mean = np.mean(sample1)

#one sample t test parameters

ttest_1samp(a = sample1, popmean = 165)
```

```python
# Q The average weight of the high-schoolers pre pandemic was 55Kg with a standard deviation of 8. Has it changed post pandemic?

#Null hypothesis: The average weight is same.
#Alternate Hypothesis: The average weight is not same.



import numpy as np
from statsmodels.stats.weightstats import ztest

# Generate a random sample
sample_1 = np.random.choice(range(30, 80), size=55)

# Perform one-sample z-test
z_statistic, p_value = ztest(sample_1, value=50)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)
```

```python
# Q Suppose you are a data analyst working for an e-commerce company, and you want to determine if there is a
# significant difference in the conversion rates of two different website designs (Design A and Design B).
# You have collected data from a sample of users who visited both designs,
# and you want to test if there is evidence that one design leads to a higher conversion rate than the other

# Null hypothesis: There is no significant difference in the conversion rate
# Alternate Hypothesis: The proportion is not same.

# Two sample z test for proportion

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Sample data
n_A = 200  # Sample size for Design A
n_B = 250  # Sample size for Design B

successes_A = 45  # Number of conversions in Design A
successes_B = 60  # Number of conversions in Design B

# Calculate sample proportions
p_A = successes_A / n_A
p_B = successes_B / n_B

# Perform two-sample Z-test for proportions
z_stat, p_value = proportions_ztest([successes_A, successes_B], [n_A, n_B], alternative='two-sided')

# Define the significance level (alpha)
alpha = 0.05

# Print results
print("Sample Proportions:")
print(f"Conversion rate for Design A: {p_A:.4f}")
print(f"Conversion rate for Design B: {p_B:.4f}")

print("\nTwo-Sample Z-Test Results:")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (significant difference)")
else:
    print("Fail to reject the null hypothesis (no significant difference)")

```

# Anova test


ANOVA or Analysis of Variance is a statistical test that compares the means or two or more groups to find significance or either groups on one another or how different they are from each other.

Analysis of Variance (ANOVA) is a statistical test used to analyze the differences among group means in a sample. ANOVA is designed to assess whether there are statistically significant differences among the means of three or more independent (unrelated) groups. It is particularly useful when you want to compare multiple groups simultaneously. ANOVA helps determine if there is more variation between the group means than would be expected due to random chanc


Types of ANOVA:

One-Way ANOVA: This is used when you have one independent variable with more than two levels or groups. It tests whether there are any significant differences between the group means.

Two-Way ANOVA: This is used when you have two independent variables, and you want to examine how they interact to affect a dependent variable. It assesses the influence of each independent variable, as well as their interaction.

Repeated Measures ANOVA: This is a variation of ANOVA used when you have repeated measurements on the same subjects or items, such as in a longitudinal study. It analyzes the impact of multiple treatments over time.



Differences from Other Tests:

Z-Test and T-Test: Z-tests and t-tests are used to compare means between two groups, whereas ANOVA is used when you have three or more groups to compare. Z-tests and t-tests are appropriate for simple pairwise comparisons, while ANOVA is suitable for simultaneously comparing multiple groups.

F-Test (F-Statistic): The F-test is a part of ANOVA used to compare variances between groups. It tells you whether the variation in the group means is larger than would be expected by random chance. In contrast, the F-test used in ANOVA is different from the F-test used in linear regression (which tests for the significance of multiple predictors).
