```python
df1 = pd.DataFrame({
    'Col1': ['a','b','c'],
    'Col2': [10,20,30]
})
df1
```

```python
df2 = pd.DataFrame({
    'Col1': ['b','c','d'],
    'Col2': [20,30,40]
})
df2
```

```python
df1.merge(df2, left_on = 'Col1',right_on = 'Col1', how = 'left')
```

```python
# df1.merge(df2, on = ['Col1','Col2'], how = 'right')
df1.merge(df2, on = 'Col1', how = 'right')
```

```python
df1.merge(df2, on = 'Col1', how = 'inner')
```

```python
df1.merge(df2, on = 'Col1', how = 'outer')
```

#### Concat 

```python
# Join two or more dataframe at row level
```

```python
df1 = pd.DataFrame({
    'Col1': ['a','b','c'],
    'Col2': [10,20,30]
})

df2 = pd.DataFrame({
    'Col1': ['d','e','f'],
    'Col3': [40,50,60]
})
```

```python
pd.concat([df1,df2])
```
